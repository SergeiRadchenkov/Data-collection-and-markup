# Задание 3 

- Определите выбросы в датасете.

- Примените технику (например, удаление выбросов,
винсоризация) для обработки выбросов.


- Выполните стандартизацию данных по выбранным
переменным.


```python
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
```

```python
z_score = stats.zscore(df['Installs'])
treshold = 3
outliers = df['Installs'][z_score > treshold]
df.loc[z_score > treshold, 'Installs'] = df['Installs'].median()
```

```python
plt.hist(df['Installs'])
plt.xlabel('Installs')
plt.ylabel('Count')
plt.show()
```

```python
z_score = stats.zscore(df['Reviews'])
treshold = 3
outliers = df['Reviews'][z_score > treshold]
df.loc[z_score > treshold, 'Reviews'] = df['Reviews'].median()
```

```python
plt.hist(df['Reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.show()
```

```python
df['Rating'] = winsorize(df['Rating'], limits=[0.1, 0.1])
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
scaler = StandardScaler()
df[['Rating']] = scaler.fit_transform(df[['Rating']])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
le = LabelEncoder()
df['Type_Encoded'] = le.fit_transform(df['Type'])
df.head()
```

```python
df = pd.get_dummies(df, columns=['Content Rating'], prefix='ContentRating', drop_first=True)
df.head()
```

```python
pivot_table = df.pivot_table(index='Category', columns='ContentRating_Teen', values='Rating', aggfunc='mean')
pivot_table
```
